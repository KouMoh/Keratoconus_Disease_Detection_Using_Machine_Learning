```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data.csv')
```

```python
if dataset.isnull().sum().sum() > 0:
    print("Missing values detected. Please handle them before proceeding.")
else:
    print("No missing values detected.")
```

```python
dataset['Keratoconus-present'] = dataset['ESI.Posterior.'].apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
print(dataset[['idEye', 'ESI.Posterior.', 'Keratoconus-present']].head())
```

```python
counts = dataset['Keratoconus-present'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Keratoconus Presence Prediction Results')
plt.xlabel('Keratoconus-present')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.show()
```
