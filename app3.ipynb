```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('data.csv')
```

```python
if dataset.isnull().sum().sum() > 0:
    print("Missing values detected. Please handle them before proceeding.")
else:
    print("No missing values detected.")
```

```python
dataset['Keratoconus-present'] = dataset['ESI.Posterior.'].apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
grouped_data = dataset.groupby(['idEye', 'Keratoconus-present']).size().unstack(fill_value=0)
```

```python
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=True, color=['blue', 'orange'], figsize=(10, 6))
plt.title('Stacked Bar Chart: Keratoconus Presence Across Eye Types')
plt.xlabel('Eye Type (idEye)')
plt.ylabel('Number of Samples')
plt.legend(title='Keratoconus-present', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
keratoconus_counts = dataset['Keratoconus-present'].value_counts()
plt.figure(figsize=(8, 8))
keratoconus_counts.plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'orange'], startangle=90)
plt.title('Pie Chart: Proportion of Keratoconus Cases')
plt.ylabel('')
plt.show()
```
